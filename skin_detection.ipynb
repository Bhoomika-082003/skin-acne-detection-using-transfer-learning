```python
import os
import shutil

# Create new folders
os.makedirs("binary_dataset/train/Acne", exist_ok=True)
os.makedirs("binary_dataset/train/Normal", exist_ok=True)

# Copy Acne images
for img in os.listdir("train/Acne"):
    shutil.copy(f"train/Acne/{img}", "binary_dataset/train/Acne")

# Copy Normal skin images
for img in os.listdir("train/Normal Skin"):
    shutil.copy(f"train/Normal Skin/{img}", "binary_dataset/train/Normal")

print("Binary dataset created successfully!")
```

```python
import os
import shutil
import random

# Create validation folders
os.makedirs("binary_dataset/val/Acne", exist_ok=True)
os.makedirs("binary_dataset/val/Normal", exist_ok=True)

def split_data(source_folder, train_folder, val_folder, split_ratio=0.8):
    images = os.listdir(source_folder)
    random.shuffle(images)

    split_index = int(len(images) * split_ratio)
    train_images = images[:split_index]
    val_images = images[split_index:]

    # Move 20% to validation
    for img in val_images:
        shutil.move(os.path.join(train_folder, img),
                    os.path.join(val_folder, img))

# Split Acne
split_data("binary_dataset/train/Acne",
           "binary_dataset/train/Acne",
           "binary_dataset/val/Acne")

# Split Normal
split_data("binary_dataset/train/Normal",
           "binary_dataset/train/Normal",
           "binary_dataset/val/Normal")

print("Validation dataset created successfully!")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
```

```python
IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    "binary_dataset/train",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary"
)

val_generator = val_datagen.flow_from_directory(
    "binary_dataset/val",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary"
)
```

```python
base_model = MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)

base_model.trainable = False  # Freeze base model

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5
)
```

```python
loss, accuracy = model.evaluate(val_generator)
print("Validation Accuracy:", accuracy)
```
